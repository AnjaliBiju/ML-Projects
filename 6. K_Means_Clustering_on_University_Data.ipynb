```python
from sklearn.cluster import KMeans
```

```python
kmeans= KMeans(n_clusters=2)
```

```python
data= df.drop('Private',axis=1)
#train every data except the Private column since that is the result we are going to cross check with
```

```python
kmeans.fit(data)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
df['Cluster']= df['Private'].apply(lambda x:1 if x=='Yes' else 0)
```

```python
df.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(df['Cluster'],kmeans.labels_))
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
#TN FP
#FN TP
```

## Conclusion

The K-Means clustering algorithm achieved an overall accuracy of ~78% when grouping universities into two clusters (Private vs Public), without using any labels during training.

* Cluster 1 (predicted as 1) aligns closely with Private universities, showing a high recall (94%), meaning most Private schools were correctly grouped.

* Cluster 0 (predicted as 0) has a low recall (35%), indicating many Public universities were misclassified as Private.

* The confusion matrix confirms this imbalance:
* 531 Private universities were correctly identified,
* but only 74 Public universities were correctly grouped, while 138 were misclassified.

This suggests that K-Means captured the pattern for Private schools much better than for Public schools, likely because Private universities have more distinct numerical characteristics (e.g., Out-of-State tuition) compared to Public ones.
